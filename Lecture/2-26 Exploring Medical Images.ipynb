# Exploring Medical Images
* load dicom files
* plot dicom images

Source: TCIA (The Cancer Imaging Archieve)

Dicom: Digital Imaging and Communication in Medicine (DICOM) format

```python
#import Packages
import imageio #image input/output
```

```python
#load the image
im = imageio.imread('tcia-chest-ct/chest-220.dcm')

#metadata
print('Image type: ',type(im))
print('Shape (in pixels): ', im.shape)
```

```python
print(im.meta.keys())
```

```python
print(im.meta) #the meta data is stored in a dictionary
```

```python
#import Packages for ploting
import matplotlib.pyplot as plt #aliasing
```

```python
#Show the image
plt.imshow(im)
plt.show()   #Do we need this line of code?
```

```python
#Show the image
plt.imshow(im, cmap = 'gray')
plt.show()   #Do we need this line of code?
```

```python
#Grey scale
plt.imshow(im, cmap = 'gray')
plt.rcParams['figure.figsize'] = [8,6] #increase size image to 8 * 6 inches
plt.show()   #Do we need this line of code?
```

```python
#For more contrast
plt.imshow(im, cmap = 'gray', vmin = -200, vmax = 200)
plt.axis('off') #turn off x and y axis
plt.show()   #Do we need this line of code?
```

```python
#import package for math stuff
import numpy as np
```

```python
#Towards 3D
im1 = imageio.imread('tcia-chest-ct/chest-220.dcm')
im2 = imageio.imread('tcia-chest-ct/chest-221.dcm')
im3 = imageio.imread('tcia-chest-ct/chest-222.dcm')

vol = np.stack([im1,im2,im3])
print('Volume dimensions: ', vol.shape)
```

```python
im4 = imageio.imread('tcia-chest-ct/chest-224.dcm')

# show images in a layout
fig, axes = plt.subplots(nrows = 2, ncols = 2)
axes[0,0].imshow(im1, cmap = 'gray', vmin = -200, vmax = 200)
axes[0,1].imshow(im2, cmap = 'gray', vmin = -200, vmax = 200)
axes[1,0].imshow(im3, cmap = 'gray', vmin = -200, vmax = 200)
axes[1,1].imshow(im4, cmap = 'gray', vmin = -200, vmax = 200)
"""[axes[0,0] axes[0,1]] 
   [axes[1,0] axes[1,1]]
   Each image is placed in a quadrant
"""
plt.show()
```

```python
# show images in a layout
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (20,20))
axes[0,0].imshow(im1, cmap = 'gray', vmin = -200, vmax = 200)
axes[0,1].imshow(im2, cmap = 'gray', vmin = -200, vmax = 200)
axes[1,0].imshow(im3, cmap = 'gray', vmin = -200, vmax = 200)
axes[1,1].imshow(im4, cmap = 'gray', vmin = -200, vmax = 200)
plt.show()
```

```python
#to load the images at once
vol = imageio.volread('tcia-chest-ct')

#metadata
print('Available metadata: ', vol.meta.keys())
print('Shape of the image array: ', vol.shape)
print('meta sampling: ', vol.meta.sampling)
```

```python
# to show all the images at once
m = vol.shape[0] #The number of images

fig, axes = plt.subplots(nrows = m, ncols= 1, figsize=(40,40))

for ii in range(m):
    im = vol[ii,:,:]
    axes[ii].imshow(im, cmap = 'gray', vmin = -200, vmax = 200)
    axes[ii].axis('off')

plt.show()

#To get rid of the scrolling go to 'Cell' -> 'Current Outputs' -> 'Toggle Scrolling'
```
